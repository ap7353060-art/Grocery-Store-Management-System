```python
# Features
1.Grocery items aur prices list me store karta hai
2.User se item quantity input leta hai
3.Automatically total bill calculate karta hai
4.Bill amount ke basis par discount apply karta hai
5.Final payable amount show karta hai
6.Simple aur easy-to-use system
```

```python
# Limitation
1.Fixed grocery items hote hain
2.Data permanently store nahi hota
3.Single customer ke liye hai
4.GST / tax calculation include nahi hai
5.Invalid input (negative value, string) handle nahi karta
```

```python
# Grocery items and prices
items = ["Milk", "Bread", "Rice", "Sugar", "Oil", "Wheat", ]
prices = [40, 30, 60, 45, 120, 78]

print("------ Grocery Store ------")
print("Items Available:")
for i in range(len(items)):
    print(items[i], ":", prices[i])

# Quantity input
quantities = []
for item in items:
    qty = int(input(f"Enter quantity of {item}: "))
    quantities.append(qty)

# Bill calculation
total_bill = 0
for i in range(len(prices)):
    total_bill += prices[i] * quantities[i]

# Discount condition
if total_bill >= 500:
    discount = total_bill * 0.10   # 10% discount
else:
    discount = 0

final_bill = total_bill - discount

# Output
print("\n------ BILL DETAILS ------")
print("Total Bill:", total_bill)
print("Discount:", discount)
print("Final Amount to Pay:", final_bill)
```
